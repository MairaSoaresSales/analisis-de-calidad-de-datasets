```python
data.WKT.str.replace('POINT', '')
data.WKT.str.split(' ', expand=True)

data['long']=data.WKT.str.split(' ', expand=True)[1]
data['lat']=data.WKT.str.split(' ', expand=True)[2]

data.long=data.long.str.strip("(")
data.lat=data.lat.str.strip(")")
```

```python
data['calle']=data.dom_establ.str.extract('([A-Za-záéíóú\.,\s]+)')
```

```python
data['altura']=data.dom_establ.str.extract('([0-9]{2,})')
```

```python
datos_utiles_url=[]
for i in data['calle']:
    try:
        datos_utiles_url.append('http://ws.usig.buenosaires.gob.ar/datos_utiles/?callback=?&calle='+i+'&altura=')
    except:
        datos_utiles_url.append(np.nan)
        print('es un nan')
```

```python
datos_utiles_url_=pd.Series(datos_utiles_url)+data['altura'].astype(str)
```

```python
altura_url=data['altura'].values.astype(str)
```

```python
def get_cp(url):
    try:
        response= requests.get(url)
        pattern='"codigo_postal":"(.*?)"'
        if response.status_code == 200:
            cp.append(re.findall(pattern, response.text)[0])
            print('.')

        return cp
    except:
        print('XXXX')
        cp.append(np.nan)
```

```python
cp=[]
for i in datos_utiles_url_:
    get_cp(i)
```

```python
cpa=[]
for i in datos_utiles_url_:
    get_cpa(i)
```
