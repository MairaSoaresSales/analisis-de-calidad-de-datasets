```python
import pandas as pd
import numpy as np
```

```python
portal=pd.read_csv('flujo-vehicular-2018-portalhastaoct.csv')
```

```python
data=pd.read_excel('Octubre - 2018.xlsx')
```

```python
data['HORA']=data.HORA.astype(int)


hora_fin=[]
for i in data['HORA']:
    if i<23:
        hora_fin.append(i+1)
    else:
        hora_fin.append(0)
        

data['hora_fin']=pd.Series(hora_fin)
data['fecha']=pd.to_datetime(data.FECHA)
data.columns=['periodo', 'ELIMINAR', 'hora', 'dia', 'estacion', 'tipo_vehiculo',
       'forma_pago', 'cantidad_pasos', 'hora_fin', 'fecha']
data=data[['periodo', 'fecha', 'hora', 'hora_fin', 'dia', 'estacion', 'tipo_vehiculo',
       'forma_pago', 'cantidad_pasos']]
```
