```python
def get_cp(url):
    try:
        response= requests.get(url)
        pattern='"codigo_postal":"(.*?)"'
        if response.status_code == 200:
            cp.append(re.findall(pattern, response.text)[0])
            print('.')

        return cp
    except:
        print('XXXX')
        cp.append(np.nan)
```

```python
cp=[]
for i in datos_utiles_url_:
    get_cp(i)
```
