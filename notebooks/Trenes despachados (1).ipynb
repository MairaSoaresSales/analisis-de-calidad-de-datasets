```python
import pandas as pd
import numpy as np
import re
import requests
from simpledbf import Dbf5
```

## Formaciones despachadas

```python
dataportal=pd.read_csv('formaciones-despachadas (1).csv', sep=';')
```

```python
data = Dbf5('sere0718.dbf', codec='Latin-1')
```

```python
data = data.to_dataframe()
```

```python
dataportal.head()
```

```python
data=data[['FR1_FECHA', 'FR1_LINEA', 'FR1_TIPO', 'FR1_REGIST', 'FR1_ORDEN', 'FR1_TREN', 'FR1_CAUC1', 'FR1_CAUC2', 'FR1_COCC1', 'FR1_COCC2', 'FR1_KM', 'FR1_KMV', 'FR1_VIAC1', 'FR1_VIAC2', 'FR1_SALC1', 'FR1_SALC2', 'FR1_SCAUC2', 'FR1_SCAUC1']]
```

```python
data
```

```python
data.FR1_FECHA=data.FR1_FECHA.apply(lambda x: pd.to_datetime(x, dayfirst=True))
```

```python
data.columns=['fr1_fecha', 'fr1_linea', 'fr1_tipo', 'fr1_regist', 'fr1_orden',
       'fr1_tren', 'fr1_cauc1', 'fr1_cauc2', 'fr1_cocc1', 'fr1_cocc2',
       'fr1_km', 'fr1_kmv', 'fr1_viac1', 'fr1_viac2', 'fr1_salc1', 'fr1_salc2',
       'fr1_scauc2', 'fr1_scauc1']
```

```python
dataportal.FR1_FECHA=dataportal.FR1_FECHA.apply(lambda x: pd.to_datetime(x, dayfirst=True))
```

```python
dataportal.columns=['fr1_fecha', 'fr1_linea', 'fr1_tipo', 'fr1_regist', 'fr1_orden',
       'fr1_tren', 'fr1_cauc1', 'fr1_cauc2', 'fr1_cocc1', 'fr1_cocc2',
       'fr1_km', 'fr1_kmv', 'fr1_viac1', 'fr1_viac2', 'fr1_salc1', 'fr1_salc2',
       'fr1_scauc2', 'fr1_scauc1']
```

```python
data2=pd.concat([dataportal,data], axis=0)
```

```python
data2
```

```python
data2.to_csv('formaciones-despachadas-subte-20181001.csv', index=False)
```

## Estado de Flota

```python
data2=pd.read_excel('Estado de la Flota_jul 2018.xlsx')
```

```python
data2
```

```python
data2['reserva']=data2['reserva'].fillna(0).astype(int)
data2['rg']=data2['rg'].fillna(0).astype(int)
data2['mantenimiento_estacionamiento']=data2['mantenimiento_estacionamiento'].fillna(0).astype(int)
data2['alistamiento']=data2['alistamiento'].fillna(0).astype(int)
```

```python
data2.columns=['mes', 'año', 'linea', 'flota', 'en_servicio', 'reserva', 'rg',
       'mantenimiento_estacionamiento', 'alistamiento']
```

```python
dataportal=pd.read_csv('estado-de-flota.csv', sep=';')
```

```python
dataportal.columns=['mes', 'año', 'linea', 'flota', 'en_servicio', 'reserva', 'rg',
       'mantenimiento_estacionamiento', 'judicial', 'alistamiento']
```

```python
dataportal=dataportal[['mes', 'año', 'linea', 'flota', 'en_servicio', 'reserva', 'rg',
       'mantenimiento_estacionamiento', 'alistamiento']]
```

```python
data3=pd.concat([dataportal, data2], axis=0)
```

```python
data3=data3.reset_index()
```

```python
data3=data3[['mes', 'año', 'linea', 'flota', 'en_servicio', 'reserva', 'rg',
       'mantenimiento_estacionamiento', 'alistamiento']]
```

```python
data3.to_csv('estado-de-flota-subte-20180912.csv', index=False)
```
