```python
def get_comuna(url):
    try:
        response= requests.get(url)
        pattern='"comuna":"(Comuna.\d+)"'
        if response.status_code == 200:
            comuna.append(re.findall(pattern, response.text)[0])
            print('.')

        return comuna
    except:
        print(url)
        comuna.append(np.nan)
```

```python
comuna=[]
for i in datos_utiles_url_:
    get_comuna(i)
```

```python
def get_barrio(url):
    try:
        response= requests.get(url)
        pattern='"barrio":"(.*?)"'
        if response.status_code == 200:
            barrio.append(re.findall(pattern, response.text)[0])
            print('.')


        return barrio
    except:
        print(url)
        barrio.append(np.nan)
        
```
