```python
import pandas as pd
import numpy as np
import re
import requests
```

```python
data=pd.read_excel('2018-10-SINIESTROS AUSA-BA Data.xlsx')
```

```python
data=data.reset_index()
```

```python
data.columns=data.loc[0,:]
```

```python
data=data.loc[1:len(data),:]
```

```python
data['año'] = pd.to_datetime(data['FECHA']).dt.year
```

```python
data['mes'] = pd.to_datetime(data['FECHA']).dt.month
```

```python
#si el mes es de dos cifras, dejar así; si es de 1 sola cifra, agregar 0 entre año y mes. HARDCODE
data['periodo']=data.año.astype(str)+data.mes.astype(str)
```

```python
data=data[['periodo', 'FECHA', 'HORA', 'AUTOPISTA', 'BANDA y/o RAMAL', 'PK',
       'CONDICIONES METEOROLÓGICAS', 'SUPERFICIE DE LA VIA', 'LESIONADOS',
       'FALLECIDOS', 'TIPO DE SINIESTRO', 'MOTO', 'LVIANO', 'BUS', 'CAMION']]
```

```python
data.columns=['periodo', 'fecha', 'hora', 'autopista', 'banda_y_o_ramal', 'pk',
       'condiciones_meteorologicas', 'superficie_de_la_via', 'lesionados',
       'fallecidos', 'tipo_de_siniestro', 'moto', 'liviano', 'bus', 'camion']
```

```python
data.fecha=data.fecha.apply(lambda x: pd.to_datetime(x, format='%d/%m/%Y'))
```

```python
data.fecha=data.fecha.dt.date
```
