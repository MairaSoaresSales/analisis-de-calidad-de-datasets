```python
def get_lat(url):
    try:
        response= requests.get(url)
        pattern='"y":."(.*?)"'
        if (re.findall(pattern, response.text)!=[]):
                lat.append(re.findall(pattern, response.text))
                print('.')
        else:
            lat.append(np.nan)
            print( i )
       
    except:
        lat.append(np.nan)
        print('sin URL')
lat=[]
for i in direccion_url:
    get_lat(i)
    print
    
lat_normalizada=[]
for i in lat:
    try:
        lat_normalizada.append((lat[lat.index(i)][0]))
    except: 
        lat_normalizada.append(np.nan)
        
data['y']=lat_normalizada
```

```python
def get_calle(url):
    try:
        response= requests.get(url)
        pattern='"nombre_calle":."(.*?)"'
        if (re.findall(pattern, response.text)!=[]):
                calle.append(re.findall(pattern, response.text))
                print('.')
        else:
            calle.append(np.nan)
            print('sin calle')
    except:
        calle.append(np.nan)
        print('sin URL')
        
calle=[]
for i in direccion_url:
    get_calle(i)
    print
    
calle_normalizada=[]
for i in calle:
    try:
        calle_normalizada.append((calle[calle.index(i)][0]))
    except: 
        calle_normalizada.append(np.nan)
```

```python
def get_altura(url):
    try:
        response= requests.get(url)
        pattern='"altura":.(\d+)'
        if (re.findall(pattern, response.text)!=[]):
                altura.append(re.findall(pattern, response.text))
                print('.')
        else:
            altura.append(np.nan)
            print('sin altura') 
    except:
        altura.append(np.nan)
        print('sin URL')
        
altura=[]
for i in direccion_url:
    get_altura(i)
    print
    
altura_normalizada=[]
for i in altura:
    try:
        altura_normalizada.append((altura[altura.index(i)][0]))
    except: 
        altura_normalizada.append(np.nan)
```

```python
def get_comuna(url):
    try:
        response= requests.get(url)
        pattern='"comuna":"(Comuna.\d+)"'
        if response.status_code == 200:
            comuna.append(re.findall(pattern, response.text)[0])
            print('.')

        return comuna
    except:
        print(url)
        comuna.append(np.nan)
        
comuna=[]
for i in datos_utiles_url:
    get_comuna(i)
    
data['COMUNA']=comuna
```
