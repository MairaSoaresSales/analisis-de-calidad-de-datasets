```python
import pandas as pd
import numpy as np
```

```python
recorridos=pd.read_csv('recorridos-realizados-20181105.CSV')
```

```python
recorridos.shape
```

## formato de tiempo de uso

```python
recorridos['bici_tiempo_uso2']=pd.to_datetime(recorridos['bici_tiempo_uso'], errors= 'coerce' )
```

```python
recorridos['bici_tiempo_uso2']=recorridos['bici_tiempo_uso2'].apply(lambda x: x - pd.Timestamp(x.date()))
```

```python
recorridos['bici_tiempo_uso2'][recorridos.bici_tiempo_uso2>'03:00:00']=np.nan
```

```python
recorridos['bici_tiempo_uso2'].mean()
```

```python
recorridos[recorridos['bici_tiempo_uso2'].isnull()]
```

```python
bici_tiempo_uso_segundos=[]
for i in recorridos['bici_tiempo_uso2']:
    bici_tiempo_uso_segundos.append(i.total_seconds())
```

```python
recorridos['bici_tiempo_uso_segundos']=pd.Series(bici_tiempo_uso_segundos)
```

## Formato fecha y hora de retiro

```python
recorridos=pd.concat([recorridos, recorridos.bici_Fecha_hora_retiro.str.split(expand=True)],axis=1)
```

```python
recorridos.columns=[             'bici_id_usuario',       'bici_Fecha_hora_retiro',
                    'bici_tiempo_uso',  'bici_nombre_estacion_origen',
               'bici_estacion_origen', 'bici_nombre_estacion_destino',
              'bici_estacion_destino',                    'bici_sexo',
                          'bici_edad',               'bici_direccion',
                          'bici_pais',             'bici_tiempo_uso2',
           'bici_tiempo_uso_segundos',                              'bici_fecha_retiro',
                                    'bici_hora_retiro']
    
```

```python
recorridos['bici_tiempo_uso2'].mean()
recorridos['bici_tiempo_uso_segundos'].mean()
recorridos.describe()
```

## obtener medias de tiempo de uso

```python
valores_medios=recorridos.groupby(['bici_fecha_retiro','bici_estacion_origen','bici_estacion_destino']).mean()['bici_tiempo_uso_segundos'].reset_index()
```

```python
valores_medios['id']=((valores_medios['bici_fecha_retiro'].astype(str))+'-'+valores_medios['bici_estacion_origen'].astype(str)+'-'+valores_medios['bici_estacion_destino'].astype(str)).str.replace('/','')
```

```python
valores_medios['id2']=valores_medios['bici_estacion_origen'].astype(str)+'-'+valores_medios['bici_estacion_destino'].astype(str).str.replace('/','')
```

```python
valores_medios_id2=valores_medios.groupby('id2').mean()['bici_tiempo_uso_segundos'].reset_index()
```

```python
print(len(valores_medios_id2.id2))
len(valores_medios_id2.id2.unique())
```

```python
print(valores_medios.shape)
len(valores_medios.id.unique())
```

```python
valores_medios.columns=['bici_fecha_retiro', 'bici_estacion_origen', 'bici_estacion_destino',
       'bici_tiempo_uso_segundosid', 'id', 'id2', 'tiempo_usoid']
```

```python
valores_medios_id2.columns=['id2', 'bici_tiempo_uso_segundosid2']
```

```python
recorridos[(recorridos['bici_estacion_origen']==87)&(recorridos['bici_estacion_destino']==502)]
```

```python
def get_hms(second):
    time = second % (24 * 3600)
    hour = time // 3600
    time %= 3600
    minutes = time // 60
    time %= 60
    seconds = time
    return("%d:%d:%d" % (hour, minutes, seconds))
```

```python
tiempo_uso=[]
for i in valores_medios['bici_tiempo_uso_segundos']:
    if i>0:
        tiempo_uso.append(get_hms(i))
    else:
        tiempo_uso.append(i)
```

```python
valores_medios['tiempo_uso']=tiempo_uso
```

```python
tiempo_uso2=[]
for i in valores_medios_id2['bici_tiempo_uso_segundosid2']:
    if i>0:
        tiempo_uso2.append(get_hms(i))
    else:
        tiempo_uso2.append(i)
```

```python
print(len(valores_medios_id2))
len(tiempo_uso2)
```

```python
valores_medios_id2['tiempo_usoid2']=tiempo_uso2
```
