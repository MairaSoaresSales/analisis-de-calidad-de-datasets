```python
def get_x(dir):
    try:
        response= requests.get(dir)
        pattern='"x":."(.\d+.\d+)"'
        
        if (re.findall(pattern, response.text)!=[]):
            x.append(re.findall(pattern, response.text)[0])
            print('.')
        else:
            x.append(np.nan)
            print( i )
    except:
        x.append(np.nan)
        print('no existe URL'+ i)
```

```python
def get_cp(url):
    try:
        response= requests.get(url)
        pattern='"codigo_postal":"(.*?)"'
        if response.status_code == 200:
            cp.append(re.findall(pattern, response.text)[0])
            print('.')

        return cp
    except:
        print(url)
        cp.append(np.nan)
```
